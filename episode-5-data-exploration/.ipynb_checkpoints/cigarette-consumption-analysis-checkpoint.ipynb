```r
library(data.table)
library(ggplot2)
#Data cleaning was completed using Open Refine. Look at doc x for more details
cigarette_data <- fread("episode-5-data-exploration/cigarette-consumption.csv")
g1 <- ggplot(cigarette_data, aes(x=Year, y=cigarettes_sold_num)) + geom_point() + geom_smooth() 
g1
```

```r
#Explore the trend over the year
cigarette_data_2018 <- cigarette_data[Year==2018,.(month_year,Year,cigarettes_sold_num)]
cigarette_data_2018$month_year <- as.Date(cigarette_data_2018[["month_year"]],format='%B-%d-%Y')
g2 <-ggplot(cigarette_data_2018, aes(x=month_year, y=cigarettes_sold_num)) + geom_point() + geom_smooth()
g2
```
